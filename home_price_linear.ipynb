```python
import pandas as pd 

import numpy as np 

import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('homeprices.csv')
```

```python
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker="+")
```

```python
new_df=df.drop('price',axis=1)
```

```python
new_df
```

```python
price=df.price
price
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(new_df,price,test_size=0.2,random_state=42)
```

```python
reg=LinearRegression()
```

```python
reg.predict(X_test)
```

```python
reg.coef_
reg.intercept_
```

```python
bestfit_point=df.area*reg.coef_+reg.intercept_
```

```python
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='blue',marker="+")

plt.plot(df.area,bestfit_point,color='red',marker='*')
```
